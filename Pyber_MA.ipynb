```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
pyber_data_df = pd.read_csv(city_data_to_load)
pyber_data2_df = pd.read_csv(ride_data_to_load)
# Combine the data into a single dataset
combined_pyber_data = pd.merge(pyber_data_df, pyber_data2_df)

# Display the data table for preview
combined_pyber_data.head()
```

```python
City_type = combined_pyber_data.groupby(["type", "city"])
avg_fare = round(City_type["fare"].mean(),2)
total_rides = (City_type["ride_id"].count())
total_drivers = City_type["driver_count"].unique()
Summary_table = pd.DataFrame({"Average Fare": avg_fare, "Total Rides": total_rides, "Total Drivers": total_drivers})
Summary_table.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
# Build the scatter plots for each city types
plt.scatter(total_rides, avg_fare, marker = "o", facecolors="Gold", edgecolors= "purple", s=total_rides*10, alpha=0.75, linewidth=1)

# Incorporate the other graph properties
plt.title("Pyber Ride Sharing Data")
plt.xlabel("Total Number of Rides (Per City)") 
plt.ylabel("Average Fare ($)")

# Create a legend
plt.legend("City Types", loc='best', fontsize='small', fancybox=True)
# Incorporate a text label regarding circle size
plt.text(70, 30, "Note: Circle size correlates with driver count per city.", verticalalignment='bottom', horizontalalignment='right', color='black', fontsize=8)

# Save Figure
plt.savefig("../Images/Pyberbubble.png")
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
#Show the 3 types of Cities
City_type = combined_pyber_data.groupby("type")
type_total_fare = City_type["fare"].sum()
type_total_rides =City_type["ride_id"].count()
type_total_drivers = [78, 490, 2405]
Summary_table = pd.DataFrame({"Total Fare": type_total_fare, "Total Rides": type_total_rides, "Total Drivers": type_total_drivers})
Summary_table
```

```python
# Calculate Type Percents
City_type_name = ["Rural", "Suburban", "Urban"]
type_total_fare = [4327.93, 19356.33, 39854.38]
colors = ("gold", "SkyBlue", "Coral")
explode = (0,0,0.1)
# Build Pie Chart
plt.title("% of Total Fares by City Type")
plt.pie(type_total_fare, explode=explode, labels=City_type_name, colors=colors, autopct="%1.1f%%", shadow=True, startangle=150)
plt.axis("equal")

# Save Figure
plt.savefig("../Images/PyberFares.png")
```
